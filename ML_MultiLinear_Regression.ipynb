# Adjusted R2 Score 

```python
# n = number of observations (data points)
# p = number of features
# Adjusted R2 calculation formula  R2 = 1 - [(1 - R2) * (n - 1) / (n - p - 1)]
```

```python
def adj_r2(y_test, y_pred, df):
    r2 = r2_score(y_test, y_pred)  # Calculates the R2 score between actual values (y_test) and predicted values (y_pred).
    n = df.shape[0]  # Number of observations - assigns the number of observations (rows) in the dataframe to the variable n.
    p = df.shape[1] - 1  # Number of independent variables (features) - assigns the number of independent variables (features) in the dataframe to the variable p. The subtraction of 1 assumes that the target variable might also be in this dataframe.
    adj_r2 = 1 - (1 - r2) * (n - 1) / (n - p - 1)  # Calculates the adj_r2 value using the adjusted R2 formula.
    return adj_r2
```

```python
adj_r2(y_test, y_pred, df)
```

```python
# The previous adjusted R2 value was close to the original R2 value because we only had 3 features. In the example below,
# let's artificially increase the number of features to see the effect on adjusted R2.
```

```python
def adj_r2(y_test, y_pred, df):
    r2 = r2_score(y_test, y_pred)
    n = df.shape[0]  # number of observations
    p = 100  # artificially set number of independent variables (features) to 100
    adj_r2 = 1 - (1 - r2) * (n - 1) / (n - p - 1)
    return adj_r2
```

```python
adj_r2(y_test, y_pred, df)
```

# Evaluation Model

```python
# At this stage, we will make an overall evaluation of the model.
```

```python
#  Calculate  various error metrics
def eval_metric(actual, pred):
    mae = mean_absolute_error(actual, pred)
    mse = mean_squared_error(actual, pred)
    rmse = np.sqrt(mean_squared_error(actual, pred))
    R2_score = r2_score(actual, pred)
    print("Model testing performance: ")
    print("---------------------------")
    print(f"R2_score \t: {R2_score}")
    print(f"MAE \t\t: {mae}")
    print(f"MSE \t\t: {mse}")
    print(f"RMSE \t\t: {rmse}")
```

```python
eval_metric(y_test, y_pred)
```
