# PCA (Principal Component Analysis)-BreastCancer

**Steps for Applying PCA**

1. **Scale the Data:**  
   - Features may be in different units and scales (e.g., temperature in °C, wind speed in km/h).  
   - **Standardization** transforms each feature to have zero mean and unit variance.  
   - This step is essential for PCA to accurately analyze variance across features.

2. **Compute the Covariance Matrix:**  
   - Used to measure the strength and direction of the relationship between features.  
   - High values in the covariance matrix indicate a strong relationship between corresponding features.

3. **Compute Eigenvalues and Eigenvectors:**  
   - Derived from the covariance matrix.  
   - **Eigenvalues:** Represent the amount of variance explained by each principal component.  
   - **Eigenvectors:** Indicate the direction of the principal components.

4. **Select Principal Components:**  
   - Sort the eigenvalues in descending order.  
   - **Variance Threshold (e.g., 80–90%):** Select the first few components that explain the desired amount of total variance.  
   - Typically, if 80% or more of the variance is explained, the chosen components are sufficient.

5. **Transform the Data onto the New Components:**  
   - The original data is projected onto the selected principal components.  
   - This transformation reduces dimensionality and concentrates the informative variance.

---

**Important Notes on PCA Usage**

1. **Suitability for Interpretability:**  
   - If you need to explain the effect of **each individual feature** (e.g., which feature contributes how much), **PCA is not suitable**.  
   - PCA represents the impact of features through components, making direct interpretation difficult.

2. **Suitability for General Evaluation:**  
   - If the focus is on a **general output or decision** rather than individual features, PCA is appropriate.  
   - For instance, in weather forecasting, if the overall forecast is more important than individual feature effects, PCA can be used.

3. **Example Scenarios:**  
   - **Not Suitable for PCA:** In customer segmentation, if explanations are required for each feature like income level, age, or purchase frequency.  
   - **Suitable for PCA:** In weather prediction where combined data from temperature, humidity, and pressure is used for a general forecast.

---

PCA is an effective tool for dimensionality reduction and variance concentration. However, its use should be determined based on the nature of the final output:
- **If feature-based interpretation is needed:** PCA is not recommended.  
- **If overall evaluation or outcome matters more:** PCA can be applied.

## PCA with sklearn

## Scaling Data

**Why Is Scaling Important in PCA?**

1. **PCA Focuses on Variance:**  
   - PCA aims to **capture the variance** in the data.  
   - Features with larger scales can **dominate** and suppress others.

2. **Standardization Should Be Preferred:**  
   - It makes the **mean zero** and **divides by the standard deviation**.  
   - Allows PCA to accurately assess variance.  
   - Min-Max scaling is **not recommended** as it doesn't preserve variance.

3. **If Not Scaled:**  
   - Different scales can **distort PCA components** and lead to misleading results.

- When applying PCA, **StandardScaler** should be used to achieve more accurate and balanced results.

```python
scaler = StandardScaler()
scaled_X = scaler.fit_transform(X)
```

```python
df_scaled_X = pd.DataFrame(scaled_X, columns = X.columns)

plt.figure(figsize = (23, 7))

sns.heatmap(df_scaled_X.corr().round(2), annot = True)
```

## PCA

**Choosing *n_components* in PCA**

The **n_components** parameter determines how many principal components will be created during PCA. When selecting this parameter, the following should be considered:

---

**Valid Values for n_components:**  
- **Number of Features and Observations:**  
  **n_components** can be at most the **minimum of the number of features or observations** in the dataset.  
  For example:  
  - 30 features, 100 observations → Maximum **30** components  
  - 100 features, 30 observations → Maximum **30** components  

---

**Factors to Consider When Choosing n_components:**

1. **Explained Variance Ratio:**  
   - Each component explains a portion of the total variance in the dataset.  
   - Typically, you aim to explain **90%–95%** of the total variance.  
     For example, if **n_components=10** explains 95% of the variance, this may be acceptable.

2. **Model Performance:**  
   - To optimize the performance of the machine learning model applied after PCA, try different **n_components** values.  
   - You can evaluate using accuracy or other performance metrics.

3. **Visualization:**  
   - If your goal is to visualize the data, **2 or 3 components** are usually sufficient to plot the data in 2D or 3D.

4. **Computational Cost:**  
   - Fewer components reduce processing time and resource usage.  
   - For large datasets, using a smaller **n_components** can help lower computational cost.

5. **Scree Plot Analysis:**  
   - Visualize the contribution of each component using a **scree plot**.  
   - Use the **elbow point** on the plot to determine the optimal number of components where added benefit begins to taper off.

6. **Domain Knowledge:**  
   - In some cases, your knowledge of the field can help determine how many components are meaningful.  
     For example, practical or theoretical knowledge may highlight a specific number of key components.

---

Choosing **n_components** in PCA is a critical step for both capturing variance and simplifying data processing:  
- **If high variance coverage is the goal:** Select enough components to explain 90–95% of the variance.  
- **For visualization purposes:** 2 or 3 components are sufficient.  
- **Using Scree Plot:** Choose the number of components at the elbow point.

## K-Means Algorithm Application

```python
# inertia hesabı

from sklearn.cluster import KMeans

ssd = []

K = range(2,10)

for k in K:
    model = KMeans(n_clusters =k, random_state=42)
    model.fit(component_df)
    ssd.append(model.inertia_)
```

```python
plt.plot(K, ssd, "bo-")
plt.xlabel("Different k values")
plt.ylabel("inertia-error") 
plt.title("elbow method")

# alttaki, grafikte sert düşüşü arıyoruz
# daha önce yaptıgımız ara işlemleri altta yapacagız tekrar
```

```python
pd.Series(ssd).diff()
```

```python
df_diff =pd.DataFrame(-pd.Series(ssd).diff()).rename(index = lambda x : x+1)
df_diff
```

```python
df_diff.plot(kind='bar')  
# nispeten 3 e daha sert düüş var, 3 olarak söyleyebilirim
```

```python
from yellowbrick.cluster import KElbowVisualizer

model_ = KMeans(random_state=42)
visualizer = KElbowVisualizer(model_, k=(2,9))

visualizer.fit(component_df)        # Fit the data to the visualizer
visualizer.show();
```

```python
from sklearn.metrics import silhouette_score

range_n_clusters = range(2,9)
for num_clusters in range_n_clusters:
    # intialise kmeans
    kmeans = KMeans(n_clusters=num_clusters, random_state=42)
    kmeans.fit(component_df)
    cluster_labels = kmeans.labels_
    # silhouette score
    silhouette_avg = silhouette_score(component_df, cluster_labels)
    print(f"For n_clusters={num_clusters}, the silhouette score is {silhouette_avg}")
```

```python
sns.scatterplot(x = component_df.first_component, y= component_df.second_component, alpha=0.7)
```

```python
model = KMeans(n_clusters =2, random_state=42)
clusters = model.fit_predict(component_df)
clusters
```
