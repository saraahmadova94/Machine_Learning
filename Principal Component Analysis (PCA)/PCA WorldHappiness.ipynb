# Principal Component Analysis (PCA)

Principal Component Analysis (PCA) is a **dimensionality reduction** and **complexity reduction** technique.

PCA is especially useful for high-dimensional datasets, as it creates a smaller number of new components (principal components) while preserving the main information — making data analysis and visualization easier.

---

## General Characteristics of PCA

1. **Dimensionality Reduction:**
   - Creates **new components** from existing features.
   - These components explain the variance in the dataset in the best way.
   - The goal is to represent most of the dataset’s information with fewer components.

2. **Feature Extraction:**
   - PCA extracts new components (**feature extraction**) instead of completely removing features like in **feature selection** (e.g., Lasso Regression).

3. **Based on Variance:**
   - PCA creates components that **maximize the variance** in the dataset.
   - For example, explaining **80% of the variance** is often considered a threshold.

4. **Solves Multicollinearity:**
   - The components created by PCA are **uncorrelated** with each other — no multicollinearity.

5. **Used in Supervised and Unsupervised Learning:**
   - PCA can be used in **supervised learning** (e.g., before regression or classification) and **unsupervised learning** (e.g., clustering).

6. **Requires Scaling:**
   - Features must be **scaled** before using PCA.
   - PCA is affected by the scale of features when measuring variance.

---

## Explaining PCA with Examples

1. **Pomegranate Example:**
   - PCA concentrates the information in a dataset.
   - Like squeezing 40kg of pomegranates to get 2kg of juice — you retain the vitamins (core components) but discard most of the mass.

2. **Bouquet Example:**
   - PCA is like summarizing a bouquet’s general characteristics (color, size) rather than identifying each flower individually.

3. **Economic Example:**
   - Instead of using hundreds of economic variables (industry, agriculture, tourism), you create a single indicator like GDP per capita.

---

## Advantages of PCA

1. **Dimensionality Reduction:**
   - Makes **data visualization** easier — especially useful for turning high-dimensional data into 2D or 3D plots.

2. **Easier EDA (Exploratory Data Analysis):**
   - Understanding key patterns becomes simpler with fewer components.

3. **Reduces Computation Time:**
   - Fewer features result in faster model training and prediction.

4. **Reduces Storage and Complexity:**
   - Less data to store, lower computational cost.

5. **Reduces Overfitting Risk:**
   - Removing unnecessary features simplifies the model and reduces the chance of overfitting.

---

## Disadvantages of PCA

1. **Information Loss:**
   - Some data is lost during dimensionality reduction — though usually within acceptable limits (e.g., ≤10%).

2. **Hard to Interpret Features:**
   - PCA components are combinations of original features, so they are harder to interpret.

3. **Only Works with Numeric Data:**
   - PCA only works with numerical data. Categorical data must be transformed first.

4. **Scaling Is Required:**
   - PCA does not work correctly if features have different scales — standardization (e.g., `StandardScaler`) is mandatory.

---

## How PCA Works

1. **Covariance Matrix:**
   - PCA builds a **covariance matrix** that describes how features vary with respect to each other.

2. **Eigenvalues and Eigenvectors:**
   - Calculates **eigenvalues** and **eigenvectors** of the covariance matrix.
     - **Eigenvalues** represent variance explained.
     - **Eigenvectors** represent the directions of new components.

3. **Creating New Components:**
   - Select components with the highest eigenvalues to create the new feature space.

4. **Explained Variance Ratio (%):**
   - Determine how much variance each component explains to decide how many components to retain.

---

## Variance and Number of Components

- **Explained Variance Ratio:**
  - Shows how well new components explain the total variance in the dataset.
  - For example, explaining ≥80% of the variance is often acceptable.
  - Components explaining 80%+ can be kept; the rest can be discarded.

- **Which Components to Select?**
  - Select components that explain the desired amount of variance (e.g., 80–90%).
  - If 10 components explain 90% of variance in a 100-feature dataset, the other 90 features can be ignored.

---

## Use Cases of PCA

1. **Supervised Learning:**
   - Improve performance of regression/classification models by reducing feature count.
   - Helps prevent **overfitting** in datasets with too many features.

2. **Unsupervised Learning:**
   - Simplify datasets for better **clustering** results (e.g., K-Means after PCA).

3. **Data Visualization:**
   - Convert high-dimensional datasets into 2D or 3D for visualization.

4. **Genetic and Biological Data:**
   - Summarize variation in gene expression data.

5. **Economics and Finance:**
   - Analyze and summarize macroeconomic indicators.

---

## Summary Principles of PCA

- **Does not eliminate features** but creates new ones based on variance.
- **Cannot handle categorical data** unless preprocessed.
- **Scaling is required** — different feature scales will negatively affect accuracy.

## **Difference Between PCA and Feature Selection**

| **Criteria**                | **PCA (Principal Component Analysis)**                               | **Feature Selection**                                          |
|----------------------------|------------------------------------------------------------------------|----------------------------------------------------------------|
| **Purpose**                | Dimensionality reduction and data summarization.                      | Selecting the most important features for the model and eliminating irrelevant ones. |
| **Approach**               | Creates new components from existing features.                        | Selects a subset of original features by evaluating importance. |
| **Output**                 | **Abstract components** derived from linear combinations of features. | A smaller subset of original features.                         |
| **Interpretation**         | Interpretation of components is difficult due to abstraction.         | Selected features are clearly related to original features.    |
| **Supervised/Unsupervised**| **Unsupervised**: Does not consider labels.                            | **Supervised**: Works based on target labels.                  
| **Scaling**                | Feature scaling is **mandatory**.                                     | Usually **not required** (depends on the algorithm).           |
| **Information Loss**       | May lose some information, but retains most variance.                 | Irrelevant features are completely removed.                    |
| **Data Type Compatibility**| Works only with **numerical data**.                                   | Works with both **numerical and categorical data**.            |
| **Example Techniques**     | PCA, Kernel PCA.                                                       | Decision Tree, Random Forest, Lasso, Recursive Feature Elimination (RFE). |

---

## **When to Use PCA vs Feature Selection**

| **Scenario**                                           | **Use PCA**                                      | **Use Feature Selection**                               |
|--------------------------------------------------------|--------------------------------------------------|----------------------------------------------------------|
| Working with a very high-dimensional dataset.          | Yes (e.g., reduce 1000 features to 10 components).| No (may be computationally expensive).                   |
| Model explainability is not a priority.                | Yes (components are hard to interpret).           | No (original features are interpretable).                |
| Model performance **and** interpretability are critical.| No (original features are lost).                  | Yes (clear view of important original features).         |
| Dimensionality reduction and visualization are needed. | Yes (e.g., reduce to 2D/3D for plotting).         | No (focus is on feature clarity, not visualization).     |

## PCA hyperparameter

## **PCA with Sklearn and Prince Libraries**

The PCA algorithm in the Scikit-learn library is a highly effective tool for data analysis. However, the **Prince** library is frequently preferred in machine learning applications due to its powerful **visualization tools**.

### **Prince PCA Parameters and Their Meanings**

1. **n_components=6**  
   - Specifies how many principal components the data will be reduced to.  
   - In this case, the data will be reduced to 6 dimensions.

2. **n_iter=10**  
   - Indicates how many iterations the PCA algorithm will perform.  
   - More iterations can lead to better results.

3. **rescale_with_mean=False**  
   - Determines whether the data will be scaled based on the mean.  
   - False: Does not scale the data using the mean.

4. **rescale_with_std=False**  
   - Determines whether the data will be scaled based on the standard deviation.  
   - False: Does not scale the data using standard deviation.

5. **copy=True**  
   - Indicates whether the PCA will make a copy of the data before applying transformations.  
   - True: The original data remains unchanged; operations are performed on a copy.

6. **check_input=True**  
   - Specifies whether to validate the input data before performing PCA.  
   - True: Input validation is performed before applying PCA.

7. **engine="sklearn"**  
   - Specifies the engine to use for PCA computation.  
   - Options include 'sklearn', 'fbpca', 'auto', etc.  
   - In this example, Scikit-learn is used.

## **1. What is an Eigenvector?**
- An eigenvector represents the **main directions** or **principal components** in your dataset.  
- Eigenvectors define the directions of the variables in the dataset, and each eigenvector represents one dimension.  
- In PCA, eigenvectors form the **axes of the principal components**, which are newly defined directions in the data space that capture the maximum variance.

**Example:**  
If your data lies in a 2D plane, the eigenvectors represent the directions along which the data spreads out the most (maximum variance).

## **2. What is an Eigenvalue?**
- An eigenvalue shows the **amount of variance** captured along the direction represented by its corresponding eigenvector.  
- The larger the eigenvalue, the more variance (information) the eigenvector explains.  
- In PCA, eigenvalues are sorted in descending order, and the eigenvector with the highest eigenvalue becomes the first principal component.

**Example:**  
If one eigenvalue is 5 and another is 1, the first eigenvector (first principal component) explains a much larger portion of the data variance than the second one.

### **PCA Process and the Role of Eigenvectors & Eigenvalues**

1. **Compute the Covariance Matrix:** Understand how the features in the dataset relate to each other.  
2. **Extract Eigenvectors and Eigenvalues:** These are computed from the covariance matrix.  
   - **Eigenvectors** determine the principal directions in the dataset.  
   - **Eigenvalues** represent how much variance is explained along those directions.  
3. **Component Selection:** Select the eigenvectors with the highest eigenvalues and project the data onto these new axes.

- **Eigenvectors:** Represent the main directions (principal components) of the data.  
- **Eigenvalues:** Indicate the amount of variance each direction explains.

```python
import prince
pca = prince.PCA(n_components=6, n_iter=10, rescale_with_mean=False,
rescale_with_std=False, copy=True, check_input=True,
engine= "sklearn" , random_state=42)

# set the hyperparameters for the PCA model:
# there were already 6 components, so we include all of them
```

```python
pca = pca.fit(df2[features])
```

```python
# eigenvalues of the model

eigeanvalues = pca.eigenvalues_
print(eigeanvalues)
```

```python
# selecting the optimal number of principal components using the Elbow method

ev = pd.DataFrame()
ev["pca"] = range(1,7)
ev["eigenvalue"]=pd.DataFrame(pca.eigenvalues_)
plt.figure(figsize=(8,6))
sns.lineplot(x="pca", y = "eigenvalue", marker = "o", data= ev)
plt.ylabel("Eigenvalue", fontsize=16)
plt.xlabel("Temel Bileşenler", fontsize=16)
plt.show()
```

```python
# Cumulative Explained Variance Ratios
# Answering the question: How much of the studied phenomenon is explained by each principal component?

explained_variance=pca.cumulative_percentage_of_variance_
print(explained_variance)
```

```python
# avo: Cumulative Explained Variance Ratio
# Main Rule: With how many principal components can we explain at least 80% of the variance?

avo = pd.DataFrame()
avo["pca"] = range(1,7)
avo["vari"] = pd.DataFrame(pca.cumulative_percentage_of_variance_)

plt.figure(figsize=(8,6))
graph = sns.barplot(x="pca", y ="vari", data=avo)

for p in graph.patches :
    graph.annotate('{:.2f}'.format(p.get_height()),(p.get_x()+0.2,p.get_height()), ha="center", va="bottom", color="black")

plt.ylabel("Açıklanan Varyans Oranı (%)", fontsize=18)
plt.xlabel("Temel Bileşenler", fontsize=18)
plt.show()  
```

```python
correlations = pca.column_correlations
correlations
```

```python
## Plot showing the positions of variables in Principal Component 1 and 2

scatter = pca.column_correlations.reset_index()

plt.figure(figsize=(10,6))

ax = sns.scatterplot(x=0, y=1, data=scatter)
ax.set(ylim=(-1,1), xlim=(-1,0))

def label_point(x, y, val, ax):
    a = pd.concat({"x": x, "y": y, "val": pd.Series(val)}, axis=1) 
    for i, point in a.iterrows():
        ax.text(point["x"]+.02, point["y"], str(point["val"]))

label_point(scatter[0], scatter[1], scatter.index, plt.gca())
plt.axvline(-0.5, ls="--")
plt.axhline(0, ls="--")
plt.title("Değişkenlerin Temel Bileşen 1 ve 2 içindeki Pozisyonları", fontsize=18)
plt.xlabel("Temel Bileşen 1 (49%)", fontsize=18)
plt.ylabel("Temel Bileşen 2 (24%)", fontsize=18)
plt.show()
```

```python
# clustering predictions based on different pairs of principal components

pd.DataFrame.iteritems = pd.DataFrame.items
import plotly.express as px
from plotly.offline import plot
components = pca.fit_transform(df2[features])
labels = {str(i): f"PC {i+1} ({var:.1f}%)"
          for i, var in enumerate(pca.cumulative_percentage_of_variance_)}
fig = px.scatter_matrix(components, labels=labels, dimensions=range(4),
                        color=df["class"])
fig.update_traces(diagonal_visible=False)
plot(fig)
```
