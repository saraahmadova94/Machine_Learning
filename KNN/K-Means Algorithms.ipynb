#### Centroid-Based Clustering
- **Method:** Data points are clustered based on their proximity to a "centroid."
- **Algorithms:**  
  - **K-Means:** Creates K clusters and assigns a centroid to each cluster.
  - **K-Means++:** Improves the K-means algorithm by selecting better initial centroids.
  - **K-Medoids:** Uses actual data points within the cluster as centroids.


#### Connectivity-Based Clustering
- **Method:** Clustering is based on the distance or connectivity between data points. Points close to each other are grouped.
- **Algorithms:**  
  - **Hierarchical Clustering:**  
    - **Agglomerative:** Starts with each data point as its own cluster; similar clusters are then merged.  
    - **Divisive:** Starts with all data points in one cluster; this cluster is then divided into smaller groups.


#### Density-Based Clustering
- **Method:** Data points are grouped based on density. Regions with higher density are treated as clusters.
- **Algorithms:**  
  - **DBSCAN (Density-Based Spatial Clustering of Applications with Noise):** Clusters dense regions and excludes noise points.  
  - **OPTICS:** Similar to DBSCAN but handles varying density levels.  
  - **HDBSCAN:** An improved and hierarchical version of DBSCAN.



#### Graph-Based Clustering
- **Method:** Represents the distances between data points as a graph and forms clusters based on the graph distances.
- **Algorithms:**  
  - **Affinity Propagation:** Identifies exemplars (cluster centers) using similarities between data points.  
  - **Spectral Clustering:** Represents data as a graph and separates groups within the graph.



#### Distribution-Based Clustering
- **Method:** Data points are clustered based on the probability of belonging to the same distribution.
- **Algorithms:**  
  - **Gaussian Mixture Models (GMM):** Each cluster is represented by a Gaussian (normal) distribution, and data points are classified based on these distributions.


#### Compression-Based Clustering
- **Method:** Transforms data into a lower-dimensional space and clusters in this compressed space.
- **Algorithms:**  
  - **BIRCH (Balanced Iterative Reducing and Clustering using Hierarchies):** A fast, scalable method for large datasets. Summarizes data into a compact representation for clustering.

# K-Means Clustering

**1. Introduction to K-Means Clustering**

K-Means clustering is a popular technique among **unsupervised learning** methods. It divides data into a predefined number of \( K \) clusters, assigning each data point to the nearest **centroid**. The algorithm aims to group data points with shared features, revealing hidden patterns and structures.

**Core Concept of K-Means:**
- **Objective:** Minimize the **variance** within clusters by assigning each data point to the nearest cluster center.
- **Optimization:** The algorithm minimizes the sum of squared distances of data points to their cluster centers (**within-cluster sum of squares, WCSS**).


**2. Why K-Means is Important**

The K-Means algorithm is vital for making sense of unlabeled data in large datasets. It is widely applicable across different fields and offers the following advantages:

1. **Simplicity and Speed:** K-Means is quick and easy to implement.
2. **Versatility in Applications:**
   - Marketing (customer segmentation)
   - Image processing (image segmentation)
   - Text analysis (document clustering)
3. **Scalability:** Suitable for handling large datasets.


**3. Understanding Clustering Algorithms**

Clustering is an **unsupervised learning** method for grouping similar data points. K-Means is among the most widely used algorithms, but other methods offer unique advantages:

**Types of Clustering Algorithms:**

1. **Hierarchical Clustering:**  
   Builds a hierarchy of clusters (e.g., agglomerative and divisive methods).
   
2. **Partition-Based Clustering:**  
   Algorithms like K-Means divide the dataset into a predefined number of clusters.

3. **Density-Based Clustering:**  
   Algorithms like DBSCAN detect clusters of arbitrary shapes using a density-based approach.


**4. How K-Means Clustering Works**

The K-Means algorithm follows these steps:

1. **Defining the Number of Clusters:**
   - The number of clusters (\( K \)) is determined.

2. **Initializing Centroids:**
   - Randomly selects \( K \) initial points as centroids, which represent the cluster centers.
   - Initial centroids can also be selected strategically using K-Means++.

3. **Assigning to Clusters:**
   - Each data point is assigned to the nearest centroid based on its distance.
   - Distance is often measured using **Euclidean Distance**.

4. **Updating Centroids:**
   - The centroid of each cluster is updated by taking the mean of all data points within the cluster.

5. **Convergence:**
   - The algorithm stops when the assignment of data points no longer changes or centroids stabilize.


**5. Applications of K-Means Clustering**

Some areas where K-Means is applied include:

1. **Customer Segmentation:**  
   Groups customers based on purchase behavior, demographics, or other attributes to inform marketing strategies.

2. **Image Segmentation:**  
   Divides an image into different regions (e.g., based on color or texture) to facilitate object recognition or image compression.

3. **Document Clustering:**  
   Groups text documents based on topics, content, or sentiment, aiding in information management.

4. **Anomaly Detection:**  
   Identifies outliers or unusual behavior (e.g., detecting anomalies in network security).


**6. Choosing the Right Number of Clusters**

Selecting the appropriate \( K \) value is crucial. The following methods can help:

1. **Elbow Method:**  
   Plots WCSS (within-cluster sum of squares) for different \( K \) values. The "elbow" point in the curve typically indicates the ideal \( K \).

2. **Silhouette Score:**  
   Compares a data point's similarity to its own cluster with its dissimilarity to other clusters. Higher scores indicate well-separated clusters.


**7. Challenges with K-Means Clustering**

1. **Sensitivity to Initialization:**  
   Poorly chosen initial centroids can lead to suboptimal results. **K-Means++** addresses this issue.

2. **Requires Predefined \( K \):**  
   The number of clusters must be specified beforehand, which may require prior knowledge of the dataset.

3. **Sensitive to Outliers:**  
   Outliers can skew centroids, leading to incorrect results.

4. **Assumes Spherical Clusters:**  
   K-Means performs best for spherical, evenly sized clusters and struggles with irregularly shaped clusters.


**8. Advanced Variants of K-Means**

1. **K-Means++:**  
   Improves centroid initialization for faster and more consistent results.

2. **Mini-Batch K-Means:**  
   Processes data in smaller batches, making it suitable for large datasets.

3. **Fuzzy C-Means:**  
   Allows data points to belong to multiple clusters with varying probabilities.

---

**9. Conclusion**

K-Means is a fundamental **clustering algorithm** in data science and machine learning. Its simplicity and efficiency make it widely used for both small and large-scale datasets. However, its limitations (e.g., sensitivity to outliers and spherical assumptions) should be considered alongside alternative algorithms. **K-Means++** and **Mini-Batch K-Means** offer practical solutions to some of these challenges.

1. **Where is K-Means clustering used?**

   Methods like the **Elbow Method**, **Silhouette Score**, and **Gap Statistics** can help identify the optimal \( K \) value.

## Hopkins test

**What is the Hopkins Test?**

The Hopkins test is a statistical method used to measure the **clustering tendency** of a dataset. It helps determine how well a dataset is suited for clustering or how randomly the data points are distributed. Performing this analysis before applying clustering algorithms like K-Means is crucial for evaluating their effectiveness.

The Hopkins test produces a value between 0 and 1:  
- **Below 0.5:** The dataset is suitable for clustering.  
- **0.3 or below:** Indicates a strong clustering tendency (ideal condition).  
- **0.5 or above:** The dataset is randomly distributed, with no clear clusters.

**Step-by-Step Application of the Hopkins Test :**

1. **Sample Selection:**  
   - Randomly select a subset of data points from the dataset. This sample forms the basis of the test.

2. **Nearest Neighbor Distances for Real Points:**  
   - Calculate the distances from each selected real data point to its nearest neighbor in the dataset.

3. **Generate Random Points:**  
   - Create a set of completely random points within the same range and dimensions as the dataset.

4. **Nearest Neighbor Distances for Random Points:**  
   - Calculate the distances from each generated random point to its nearest neighbor in the dataset.

5. **Calculate the Hopkins Score:**  
   - Compare the sum of nearest neighbor distances for real points (\( W \)) with that of random points (\( U \)).

**Interpreting Results**

- **\( H \) < 0.5:** The dataset is suitable for clustering.  
- **\( H \) \(\approx\) 0.3 or lower:** Indicates a strong clustering tendency (ideal condition).  
- **\( H \) ≥ 0.5:** The dataset is randomly distributed, with no apparent clustering structure.

**Why is the Hopkins Test Important?**

The Hopkins test is a quick and effective method to determine whether your dataset is clusterable. The test result helps assess the applicability of clustering algorithms like K-Means, saving time and resources by avoiding unnecessary attempts on unsuitable datasets.

```python
from sklearn.neighbors import BallTree  # BallTree is a data structure used to efficiently perform distance queries between data points. 
import numpy as np
import pandas as pd
def hopkins(data_frame, sampling_size):
    """Assess the clusterability of a dataset. A score between 0 and 1, a score around 0.5 express
    no clusterability and a score tending to 0 express a high cluster tendency.
    Parameters
    ----------
    data_frame : numpy array
        The input dataset
    sampling_size : int
        The sampling size which is used to evaluate the number of DataFrame.
    Returns
    ---------------------
    score : float
        The hopkins score of the dataset (between 0 and 1)
    Examples
    --------
    >>> from sklearn import datasets
    >>> from pyclustertend import hopkins
    >>> X = datasets.load_iris().data
    >>> hopkins(X,150)
    0.16
    """
    if type(data_frame) == np.ndarray:
        data_frame = pd.DataFrame(data_frame)
    # Sample n observations from D : P
    if sampling_size > data_frame.shape[0]:
        raise Exception(
            'The number of sample of sample is bigger than the shape of D')
    data_frame_sample = data_frame.sample(n=sampling_size)
    # Get the distance to their neirest neighbors in D : X
    tree = BallTree(data_frame, leaf_size=2)
    dist, _ = tree.query(data_frame_sample, k=2)
    data_frame_sample_distances_to_nearest_neighbours = dist[:, 1]
    # Randomly simulate n points with the same variation as in D : Q.
    max_data_frame = data_frame.max()
    min_data_frame = data_frame.min()
    uniformly_selected_values_0 = np.random.uniform(min_data_frame[0], max_data_frame[0], sampling_size)
    uniformly_selected_values_1 = np.random.uniform(min_data_frame[1], max_data_frame[1], sampling_size)
    uniformly_selected_observations = np.column_stack((uniformly_selected_values_0, uniformly_selected_values_1))
    if len(max_data_frame) >= 2:
        for i in range(2, len(max_data_frame)):
            uniformly_selected_values_i = np.random.uniform(min_data_frame[i], max_data_frame[i], sampling_size)
            to_stack = (uniformly_selected_observations, uniformly_selected_values_i)
            uniformly_selected_observations = np.column_stack(to_stack)
    uniformly_selected_observations_df = pd.DataFrame(uniformly_selected_observations)
    # Get the distance to their neirest neighbors in D : Y
    tree = BallTree(data_frame, leaf_size=2)
    dist, _ = tree.query(uniformly_selected_observations_df, k=1)
    uniformly_df_distances_to_nearest_neighbours = dist
    # return the hopkins score
    x = sum(data_frame_sample_distances_to_nearest_neighbours)
    y = sum(uniformly_df_distances_to_nearest_neighbours)
    if x + y == 0:
        raise Exception('The denominator of the hopkins statistics is null')
    return x / (x + y)[0]
```

## Silhouette analysis

**What is Silhouette Analysis?**

**Silhouette analysis** is a metric used to evaluate the performance of clustering. By comparing **cohesion** (similarity within clusters) and **separation** (similarity between clusters), it shows how well clusters are separated and whether the data points are correctly clustered. A **silhouette value** is calculated for each data point, and the average of these values gives an overall measure of clustering quality.

The **Silhouette Score** is typically calculated using the **Minkowski distance** or **Euclidean distance** and returns a value between **[-1, 1]**.

**Silhouette Value Ranges and Interpretations**

- **1:** The data point is well-placed in its cluster and is completely isolated from other clusters (Good clustering).  
- **0:** The data point is near the boundary between two clusters (Uncertain clustering).  
- **-1:** The data point is misclassified and might belong to another cluster (Poor clustering).

**Practical Silhouette Values**

Although there are no strict thresholds, the following interpretations are common in practice:
- **Greater than 0.5:** Indicates good clustering performance.  
- **Between 0.25 and 0.5:** Clustering results can be improved.  
- **Less than 0.25:** The clustering algorithm may not be working well for the dataset.


**Calculating the Silhouette Score**

The Silhouette Score is based on two key metrics:

1. **Cohesion:**  
   - The average distance of a data point to all other points in its cluster.  
   - **Low cohesion** indicates that the point is close to others in its cluster.

2. **Separation:**  
   - The minimum average distance of a data point to points in other clusters.  
   - **High separation** means the point is far from other clusters.

**Formula:**

s(i) = \frac{b(i) - a(i)}{\max(a(i), b(i))}

- \(a(i)\): The average distance of the point to other points in its cluster (**cohesion**).  
- \(b(i)\): The minimum average distance of the point to points in the nearest neighboring cluster (**separation**).


**Uses of Silhouette Analysis**

1. **Evaluating Clustering Quality:**  
   - It measures the success of a clustering algorithm.
2. **Determining the Optimal Number of Clusters:**  
   - Calculate the silhouette score for different cluster counts and choose the \(k\) that gives the highest average score.
3. **Improving the Clustering Model:**  
   - A low silhouette score indicates that improvements in the clustering algorithm or preprocessing steps are needed.


**Key Criteria for Successful Clustering**

1. **Data points within a cluster should be close to each other:**  
   - The distances between points in a cluster should be **minimized**.  
   - This represents **cohesion** and increases the similarity within the cluster.

2. **Clusters should be far from each other:**  
   - The distances between the centers of different clusters should be **maximized**.  
   - This represents **separation** and ensures clear distinctions between clusters.

These two criteria are essential for **improving clustering performance** and **producing meaningful results**.

```python
from sklearn.metrics import silhouette_score
```

```python
silhouette_score(X, k_means_model.labels_) # k=5
```

```python
range_n_clusters = range(2,9)
for num_clusters in range_n_clusters:
    # intialise kmeans
    kmeans = KMeans(n_clusters=num_clusters, random_state=42)
    kmeans.fit(X)
    cluster_labels = kmeans.labels_
    # silhouette score
    silhouette_avg = silhouette_score(X, cluster_labels)
    print(f"For n_clusters={num_clusters}, the silhouette score is {silhouette_avg}")
```

```python
# create a graph that shows both the Elbow Method and the Silhouette Score

inertia_scores = []
silhouette_scores = []
for i in range(2, 10):
    model = KMeans(n_clusters=i, random_state=42)
    labels = model.fit_predict(X)
    silhouette = silhouette_score(X, labels)
    inertia = model.inertia_
    inertia_scores.append(inertia)
    silhouette_scores.append(silhouette)
# Plotting the Elbow Method with Inertias and Silhouette Scores
plt.figure(figsize=(10, 6))
# Plot Inertias
plt.plot(range(2,10), inertia_scores, label='Inertia', marker='o')
# Plot Silhouette Scores on a secondary y-axis
plt.twinx()
plt.plot(range(2,10), silhouette_scores, label='Silhouette Score', marker='o', color='orange')
# Add labels, title, and legend
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.title('Elbow Method with Inertias and Silhouette Scores')
plt.legend(loc='upper left', bbox_to_anchor=(0.7, 1))
plt.show()
```

## Adjust Rand Score

**What is Rand Index (Rand Score)?**

**Rand Index** is a metric used to evaluate the performance of a clustering algorithm. Specifically, it measures **how similar** two clustering results are—typically one set of predicted clusters from the algorithm and another set of ground truth labels. This metric is ideal for assessing the accuracy of clustering results based on correct classifications.

**Rand Index Value**

The Rand Index takes a value between **0 and 1**:
- **1:** Indicates a perfect match; the predicted clusters align completely with the true labels.
- **0:** Indicates a random match; there is no alignment between the predicted clusters and true labels.

**Interpretation of Rand Index Values**

| **Rand Index Range**    | **Interpretation**      |
|--------------------------|-------------------------|
| **0.1 or below**         | No agreement, weak fit  |
| **0.1 - 0.3**            | Low agreement           |
| **0.3 - 0.5**            | Moderate agreement      |
| **0.5 - 1.0**            | Good agreement, meaningful results |

**Calculating Rand Index**

The Rand Index is calculated as the ratio of correct classifications to the total number of examples. Correct classifications include:
1. **True Positive (TP):** Pairs correctly classified as belonging to the same cluster.
2. **True Negative (TN):** Pairs correctly classified as belonging to different clusters.

**Formula:**

RI = \frac{TP + TN}{TP + TN + FP + FN}

Where:
- **TP:** True Positive  
- **TN:** True Negative  
- **FP:** False Positive (pairs incorrectly grouped together)  
- **FN:** False Negative (pairs incorrectly separated)

**Key Points**

- **High Rand Index (0.5 - 1.0):** Indicates the clustering algorithm is grouping data close to the actual classes.  
- **Low Rand Index (0.1 - 0.3):** Indicates the clustering algorithm is far from the true classes.  
- **Rand Index is especially useful for evaluating the performance of unsupervised learning algorithms.**

## Hopkins test & Elbow method & Silhouette analysis

```python
hopkins(X2, X2.shape[0]) # for 4 features it was 0.16
```

```python
ssd = []
K = range(2,10)
for k in K:
    model3 = KMeans(n_clusters =k)
    model3.fit(X2)
    ssd.append(model3.inertia_)
```

```python
plt.plot(K, ssd, "bo-")
plt.xlabel("Different k values")
plt.ylabel("inertia-error")
plt.title("elbow method")
```

```python
df_diff = pd.DataFrame(-pd.Series(ssd).diff()).rename(index = lambda x : x+1)
df_diff
```

```python
df_diff.plot(kind='bar');
```

```python
from yellowbrick.cluster import KElbowVisualizer
model_ = KMeans(random_state=42)
visualizer = KElbowVisualizer(model_, k=(2,9))
visualizer.fit(X2)        # Fit the data to the visualizer
visualizer.show();
```

```python
range_n_clusters = range(2,9)
for num_clusters in range_n_clusters:
    # intialise kmeans
    kmeans = KMeans(n_clusters=num_clusters, random_state=42)
    kmeans.fit(X2)
    cluster_labels = kmeans.labels_
    # silhouette score
    silhouette_avg = silhouette_score(X2, cluster_labels)
    print(f"For n_clusters={num_clusters}, the silhouette score is {silhouette_avg}")
```

```python
# Elbow + Silhoutee skorunu bir arada gösteren grafik

inertia_scores = []
silhouette_scores = []
for i in range(2, 10):
    model = KMeans(n_clusters=i, random_state=42)
    labels = model.fit_predict(X2)
    silhouette = silhouette_score(X2, labels)
    inertia = model.inertia_
    inertia_scores.append(inertia)
    silhouette_scores.append(silhouette)
    
# Plotting the Elbow Method with Inertias and Silhouette Scores
plt.figure(figsize=(10, 6))

# Plot Inertias
plt.plot(range(2,10), inertia_scores, label='Inertia', marker='o')

# Plot Silhouette Scores on a secondary y-axis
plt.twinx()
plt.plot(range(2,10), silhouette_scores, label='Silhouette Score', marker='o', color='orange')

# Add labels, title, and legend
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.title('Elbow Method with Inertias and Silhouette Scores')
plt.legend(loc='upper left', bbox_to_anchor=(0.7, 1))
plt.show()
```
