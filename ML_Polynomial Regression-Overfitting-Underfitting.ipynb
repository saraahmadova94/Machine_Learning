## Evaluation on the Test Set

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
```

```python
y_pred = model_poly.predict(X_test)   
```

```python
def eval_metric(actual, pred):
    mae = mean_absolute_error(actual, pred)
    mse = mean_squared_error(actual, pred)
    rmse = np.sqrt(mean_squared_error(actual, pred))
    R2_score = r2_score(actual, pred)
    print("Model testing performance:")
    print("--------------------------")
    print(f"R2_score \t: {R2_score}")
    print(f"MAE \t\t: {mae}")
    print(f"MSE \t\t: {mse}")
    print(f"RMSE \t\t: {rmse}")
```

```python
eval_metric(y_test, y_pred)  # The values of the test section
```

# Find an Optimum Polynomial Degree

```python
# We will decide the optimal degree based on the RMSE value
```

```python
def poly(d):  # Depending on what writeS for the degree, it will generate a solution

    train_rmse_errors = []   # Will store the RMSE values for the training set
    test_rmse_errors = []    # Will store the RMSE values for the test set
    number_of_features = []  # Will track how many features are created as the degree increases

    for i in range(1, d):  # Will iterate from 1 to the desired degree

        polynomial_converter = PolynomialFeatures(degree = i, include_bias =False)
        poly_features = polynomial_converter.fit_transform(X)
        
        X_train, X_test, y_train, y_test = train_test_split(poly_features, y, test_size=0.3, random_state=101)
        
        model = LinearRegression(fit_intercept=True) 
        model.fit(X_train, y_train)
        
        train_pred = model.predict(X_train)
        test_pred = model.predict(X_test)
        
        train_RMSE = np.sqrt(mean_squared_error(y_train,train_pred))
        test_RMSE = np.sqrt(mean_squared_error(y_test,test_pred))
        
        train_rmse_errors.append(train_RMSE)
        test_rmse_errors.append(test_RMSE)
        
        number_of_features.append(poly_features.shape[1])
        
    return pd.DataFrame({"train_rmse_errors": train_rmse_errors, "test_rmse_errors":test_rmse_errors, "number of features":number_of_features}, 
                        index=range(1,d))
```

```python
poly(10)
```

```python
# Let's plot the results above to observe the explosion point

plt.plot(range(1,10), poly(10)["train_rmse_errors"], label = "TRAIN")
plt.plot(range(1,10), poly(10)["test_rmse_errors"], label = "TEST")
plt.xlabel("Polynamial Complex")
plt.ylabel("RMSE")
plt.legend();
plt.show()
```

```python
plt.plot(range(1,6), poly(6)["train_rmse_errors"], label = "TRAIN")
plt.plot(range(1,6), poly(6)["test_rmse_errors"], label = "TEST")
plt.xlabel("Polynamial Complex")
plt.ylabel("RMSE")
plt.legend();
plt.xticks(range(1, 6, 1))
plt.show()
```

```python
# Staying at the value just before the explosion point will be more reliable.
# Therefore, the 3rd degree can be taken as the optimal value.
```

# Overfitting Uygulama Denemesi

```python
# Let's generate features based on the 5th degree, which is the point where overfitting starts in the graph above
# (the point where train and test diverge), and observe the deliberate formation of overfitting.
```

```python
over_poly_converter = PolynomialFeatures(degree=5, include_bias=False)
# Generating 5th-degree polynomial features
```

```python
over_model =LinearRegression()   
```

```python
X_train, X_test, y_train, y_test = train_test_split(over_poly_converter.fit_transform(X), y, test_size=0.3, random_state=101)
```

```python
over_model.fit(X_train, y_train)
```

```python
y_pred_over = over_model.predict(X_test)
```

```python
eval_metric(y_test, y_pred_over)
```

```python
y_train_over = over_model.predict(X_train)
```

```python
eval_metric(y_train, y_train_over)
```

```python
# The significant difference between the train R2 (0.99) and test R2 (0.76) shows overfitting.
# Despite the train set having excellent scores, the poor performance of the test set indicates overfitting.
```
